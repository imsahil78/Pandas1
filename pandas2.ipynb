# Merging Joining Concat

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'inner') # on = kiske bases pe merge
                                                    # how = 'inner' employee_id pe common parts pe merge
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'outer')  # how = "outer" common and uncommon
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'left')  # how = "left" merge based on employees(left ke sare cell)
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'right')  # how = "right" merge based on employees(left ke sare cell)
```

# GroupBy 

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

```python
# Group by Category and calculate the sum of Sales
cat = df.groupby('Category')['Sales'].sum() 
cat
```

```python
# Group by Store and calculate the sum of Sales
cat = df.groupby('Store')['Sales'].sum()
cat
```

```python
# Group by multiple columns
# Group by Category and Store
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

# Aggregation

```python
#multiple value to combine krke value

print("mean: ",df['Sales'].mean())
print("median: ",df['Sales'].median())
print("max: ",df['Sales'].max())
print("min: ",df['Sales'].min())
print("count: ",df['Sales'].count())
print("std: ",df['Sales'].std())
```

```python
df['Sales'].agg(['mean', 'median', 'max', 'min', 'count'])
```
